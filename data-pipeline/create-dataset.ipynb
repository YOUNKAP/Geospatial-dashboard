## Creating Additional Datasets
> Creating additional datasets from 
1. Transmission Lines
2. Electric Substations

```python
import geopandas as gpd
from shapely.geometry import Point


trx = gpd.read_file("transmission_lines.csv")
es = gpd.read_file("electric_substations.csv")


def processDatasets(file_name, trl):
    print(f'Processing {file_name}')
    for idx, row in trl.iterrows():
        trl.at[idx , 'geometry'] = Point((float(row["longitude"]), float(row["latitude"])))

    trl = trl[["geometry"]]

    display(trl)
    trl.to_file(f"./{file_name}.geojson", driver='GeoJSON')

    with open(f'./{file_name}.geojson','r') as file:
        data = file.read().replace('\n', "").replace(" ","")
        with open(f'./{file_name}.geojson','w') as fw:
            fw.write(data)
            fw.close()
    print("Done!")
    

processDatasets("transmission_lines", trx)
processDatasets("electrical_substations",es)

```
